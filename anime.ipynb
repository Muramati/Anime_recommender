## Data cleaning

```python
class DataUnderstanding:
    def __init__(self, df):
        self.df = df
    
    def get_missing_values(self):
        missing_values = self.df.isnull().sum()
        return missing_values
    
    def get_dtypes(self):
        dtypes = self.df.dtypes
        return dtypes
    
    def get_value_counts(self):
        value_counts = {}
        for column in self.df.columns:
            value_counts[column] = self.df[column].value_counts()
        return value_counts
    
    def set_index(self, column):
        self.df.set_index(column, inplace=True)

```

## Preparing DATA for modelling

```python
#relevant features for recommendation
selected_features = ['genre']
```

```python
#replace null values in feature with string 
for feature in selected_features:
    final_dataset[feature] = final_dataset[feature].fillna('')
```

```python
combined_features = final_dataset['genre']
```

```python
print(combined_features)
```

```python
#converting text data to feature vectors
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
#using cosine similarity for the similarity scores
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
print(similarity.shape)
```

```python
#get anime name from user
anime_name = input('Enter Anime you have watched')
```

```python
#create a list with all animes in the final dataseta
anime_list = final_dataset['title'].tolist()

```

```python
import difflib
```

```python
# finding the close match for the anime name given by the user
find_close_match = difflib.get_close_matches(anime_name,anime_list)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
#find index of movie with title
index_of_anime = final_dataset[final_dataset.title == close_match]['index'].values[0]
```

```python
print(index_of_anime)
```

```python
#getting a list of similar animes
similarity_score = list(enumerate(similarity[index_of_anime]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sort animes based on similarity score
sorted_similar_animes = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_animes)
```

```python
final_dataset.index = final_dataset.index.astype(float)
```

```python
type(final_dataset.index)
```

```python
# print the name of similar animes based on the index

print('Animes suggested for you : \n')

i = 0

for anime in sorted_similar_animes:
  index = anime[0]
  title_from_index = final_dataset[final_dataset.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```

Anime Recomender System

```python
anime_name = input('Enter Anime you have watched')
anime_list = final_dataset['title'].tolist()
find_close_match = difflib.get_close_matches(anime_name,anime_list)
close_match = find_close_match[0]
index_of_anime = final_dataset[final_dataset.title == close_match]['index'].values[0]
similarity_score = list(enumerate(similarity[index_of_anime]))
sorted_similar_animes = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print('Animes suggested for you : \n')

i = 0

for anime in sorted_similar_animes:
  index = anime[0]
  title_from_index = final_dataset[final_dataset.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
